```python
# importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# loading dataset
df=pd.read_csv('data.csv')
```

```python
# getting description of dataset
df.describe()
```

```python
df.drop(['id', 'Unnamed: 32'], axis=1, inplace=True)
```

# Exploratory Data Analysis

```python
def convertToNumeric(value):
    if(value=='M'):
        return 1
    return 0
```

```python
df['diagnosis'] = df['diagnosis'].apply(convertToNumeric)
```

```python
plt.figure(figsize=(10,6))
df[df['diagnosis'] == 1]['radius_mean'].plot(kind='hist',alpha=0.5, color = 'blue', label='Malignant')
df[df['diagnosis'] == 0]['radius_mean'].plot(kind='hist',alpha=0.5, color = 'red', label='Benign')
plt.legend()
```

```python
sns.barplot(data=df, x='diagnosis', y='radius_mean', hue = 'diagnosis' , palette='Set1')
plt.title('Radius Mean vs Diagnosis')
```

```python
sns.countplot(data=df, x='diagnosis', hue='diagnosis', palette='Set2')
plt.title('Count Plot of Diagnosis')
```

```python
sns.heatmap(df[['radius_mean', 'perimeter_mean', 'area_mean']].corr(), annot=True)
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('diagnosis', axis=1)
y = df['diagnosis']
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=101)
```

# Standardization

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

# Support Vector Machine(SVM)

```python
from sklearn.svm import SVC
```

```python
svc_model = SVC()
```

```python
svc_model.fit(X_train_scaled, y_train)
```

```python
preds = svc_model.predict(X_test_scaled)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test, preds))
```

```python
print(classification_report(y_test, preds))
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(confusion_matrix(y_test, preds), annot=True, fmt='d', cmap='Greens')
plt.title('Confusion Matrix for SVM Model')
plt.xlabel('Predicted')  # Add this
plt.ylabel('Actual')
```

# Gridsearch

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C': [0.1,1, 10, 100, 1000], 'gamma': [1,0.1,0.01,0.001,0.0001], 'kernel': ['rbf']}
```

```python
grid = GridSearchCV(SVC(), param_grid, verbose=3)
```

```python
grid.fit(X_train_scaled, y_train)
```

```python
print("Best Parameters: ", grid.best_params_)
```

```python
print("Best Estimators: ",grid.best_estimator_)
```

```python
grid_preds=grid.predict(X_test_scaled)
```

```python
print(confusion_matrix(y_test, grid_preds))
```

```python
print(classification_report(y_test, grid_preds))
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(confusion_matrix(y_test, grid_preds), annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix for Grid Search Model')
plt.xlabel('Predicted')  # Add this
plt.ylabel('Actual')
```

# Conclusion
* Baseline SVM model with default parameters: 98% accuracy (3 misclassifications)
* After hyperparameter tuning using Grid Search with 5-fold cross-validation
* Best parameters found: C=10, gamma=0.001, kernel='rbf'
 * Optimized model: 96% accuracy (6 misclassifications)
 * Baseline model performed better on test set, demonstrating that default parameters were already well-suited for this dataset
 * Both models achieved high accuracy (>96%), showing SVM's effectiveness for high-dimensional medical data
